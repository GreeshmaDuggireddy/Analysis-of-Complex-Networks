### Dataset  1 - Word Adjacencies

```python
wordadj = nx.read_gml("adjnoun.gml")
nx.draw(wordadj, with_labels=True)
```

```python
print(nx.info(wordadj))
```

```python
nx.shortest_path_length(wordadj, source='family')
```

```python
wordadj_avg_path_length = nx.average_shortest_path_length(wordadj)
wordadj_avg_path_length
```

```python
wordadj_diameter = nx.diameter(wordadj)
wordadj_diameter
```

```python
nx.number_of_nodes(wordadj)
```

```python
x = [5,25,125,15625,244140625]
y = [112,12544,157351936,24759631762948096,613039365036788240314949190025216]
 
fig = plt.figure(figsize=(8, 6))
plt.scatter(x,y)
plt.plot(x,y)
 
plt.loglog(basex=10,basey=2)
plt.xlabel('Diameter')
plt.ylabel('Number of Nodes')
 
plt.show()
```

```python
print(nx.clustering(wordadj))
```

```python
print(nx.average_clustering(wordadj))
```

```python
import matplotlib.pyplot as plt

# Implement ---
for N in [112]:
    degree_dictionary = dict(wordadj.degree())
    D = [degree_dictionary[i] for i in degree_dictionary.keys()]
    Sorted_Degrees = np.sort(D)
    Arranged = np.arange(1., N+1)
    CCDF = 1 - Arranged/N 
    #Plotting the graph on a loglog scale.
    plt.loglog(Sorted_Degrees,CCDF) 
plt.xlabel('Degree')
plt.ylabel('CCDF')
plt.plot() 
# ---
```

```python
nx.degree_centrality(wordadj)
```

```python
nx.betweenness_centrality(wordadj)
```

```python
nx.closeness_centrality(wordadj)
```

```python
nx.eigenvector_centrality(wordadj)
```

### Dataset 5 - Political Blogs Dataset

```python
polblogs = nx.read_gml("polblogs.gml")
print(type(polblogs))             
print(polblogs.edges(data=True))
```

```python
nx.nodes(polblogs)
```

```python
nx.shortest_path_length(polblogs, source="blog.veggiedude.com")
```

```python
largest_polblogs = polblogs.subgraph(sorted(nx.connected_components(polblogs), key=len, reverse=True)[0])
```

```python
nx.info(largest_polblogs)
```

```python
nx.number_of_nodes(largest_polblogs)
```

```python
x = [8,64,512,262144,68719476736,4722366482869645213696]
y = [1222,1493284,2229897104656,4972441097353211816878336,24725170466647213317508510938331679164007826128896,611334054604763576362945515790513134547743186000322525906475461492192558152675224294181496806178816]
 
fig = plt.figure(figsize=(8, 6))
plt.scatter(x,y)
plt.plot(x,y)
 
plt.loglog(basex=10,basey=10)
plt.xlabel('Diameter')
plt.ylabel('Number of Nodes')
plt.show()
```

```python
largest_polblogs=largest_polblogs.to_undirected()
```

```python
nx.diameter(largest_polblogs)
```

```python
nx.info(largest_polblogs)
```

```python
polblogs_avg_path_length = nx.average_shortest_path_length(largest_polblogs)
polblogs_avg_path_length
```

```python
G = nx.DiGraph()
for u,v in largest_polblogs.edges():
    if G.has_edge(u,v):
        G[u][v]['weight'] += 1
    else:
        G.add_edge(u, v, weight=1)
```

```python
nx.info(G)
```

```python
print(nx.clustering(G))
```

```python
print(nx.average_clustering(G))
```

```python
import matplotlib.pyplot as plt

# Implement ---
for N in [1222]:
    degree_dictionary = dict(G.degree())
    D = [degree_dictionary[i] for i in degree_dictionary.keys()]
    Sorted_Degrees = np.sort(D)
    Arranged = np.arange(1., N+1)
    CCDF = 1 - Arranged/N 
    #Plotting the graph on a loglog scale.
    plt.loglog(Sorted_Degrees,CCDF) 
plt.xlabel('Degree')
plt.ylabel('CCDF')
plt.plot() 
# ---
```

```python
nx.degree_centrality(G)
```

```python
nx.betweenness_centrality(G)
```

```python
nx.closeness_centrality(G)
```

```python
nx.eigenvector_centrality(G,max_iter=500)
```

```python
blogs = G
```

```python
nx.write_gexf(blogs, "blogs.gexf")
```
